```python
# %load bad.py
%matplotlib inline
from numpy import *
import matplotlib.pyplot as plt

T = 100
def lif(T):
    """Leaky integrate and fire neuron
    
    Params
    ------
    T : num time steps
    
    Returns
    ------
    Times : the time steps for the simulation
    Vs : voltages for each step
    """
    # Initialize time
    dt = 0.125 # time step
    times = arange(0, T + dt, dt)
    time_refracr = 0
    
    # Initialize LIF parameters
    Vs = zeros(len(times))       # potential (V) trace over time
    Rm = 1                      # Rawr! Resistance! (kOhm)
    Cm = 10                     # capacitance (uF)
    tau_m = Rm * Cm             # time constant (msec)
    tau_refractory = 4          # refractory period (msec)
    Vth = 1                     # Threshold
    spike = 0.5                 # spike (V)
    I = 1.5                     # input current (A)
    
    # Starting simulation
    for i, t in enumerate(times):
        # Are we in refractory period,
        # if not, then update the voltage
        if t > time_refracr:
            
            # Integrating the equation
            Vs[i] = Vs[i - 1] + (-Vs[i - 1] + I * Rm) / tau_m * dt
            if Vs[i] >= Vth:
                Vs[i] += spike
                time_refracr = t + tau_refractory
    return Vs, times

def plot_lif(times, Vs):
    # Run and plot membrane potential trace
    plt.plot(times, Vs)
    plt.title('Leaky Integrate-and-Fire Example')
    plt.ylabel('Membrane Potential (V)')
    plt.xlabel('Time (msec)')
    plt.ylim([0, 2])
    plt.show()
    plt.close()
```

```python
times, Vs = lif(100)
plot_lif(times, Vs)
```
